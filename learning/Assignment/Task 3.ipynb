""Problem: Basic Calculator""
Objective: Create a simple calculator program that can perform basic arithmetic operations: addition, subtraction, multiplication, and division.


```python
def add(x, y):
    return x + y
def subtract(x, y):
    return x - y
def multiply(x, y):
    return x * y
def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Error: Division by zero"
    
    """
    param 1: x (float or int)
    param 2: y (float or int)
    Return the sum of x and y.
    type_return: float or int
    Return the subtraction of y from x.
    type_return: float or int
    Return the multiplication of x and y.
    Return the division of x by y.
    If y = 0, return error message.
    """
    
while True:
    print("\n=== Basic Calculator ===")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")

    choice = input("Enter choice (1/2/3/4/5): ")

    if choice == '5':
        print("Thank you for using the calculator. Goodbye!")
        break

    if choice not in ['1', '2', '3', '4']:
        print("Invalid choice. Please try again.")
        continue

    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
    except ValueError:
        print("Invalid input. Please enter numeric values.")
        continue

    if choice == '1':
        result = add(num1, num2)
        operation = "adding"
    elif choice == '2':
        result = subtract(num1, num2)
        operation = "subtracting"
    elif choice == '3':
        result = multiply(num1, num2)
        operation = "multiplying"
    elif choice == '4':
        result = divide(num1, num2)
        operation = "dividing"

    print(f"The result of {operation} {num1} and {num2} is {result}")

    again = input("Do you want to perform another calculation? (yes/no): ").lower()
    if again != "yes":
        print("Thank you for using the calculator. Goodbye!")
        break
```

```python
def calculate(x, y, operator):
    """Perform basic arithmetic operations."""
    if operator == "+":
        return x + y
    elif operator == "-":
        return x - y
    elif operator == "*":
        return x * y
    elif operator == "/":
        return x / y if y != 0 else "Error: Division by zero"

print("\n=== Welcome to the Basic Calculator ===")
print("Addition (+) | Subtraction (-) | Multiplication (*) | Division (/)")
print("Type 'exit' to quit.\n")

while True:
    try:
        num1 = float(input("Enter the first number: "))
    except ValueError:
        print("Invalid input. Please enter a numeric value.")
        continue

    while True:
        choice = input("Enter an operation (+, -, *, /): ").strip()
        if choice in ["+", "-", "*", "/"]:
            break
        print("Invalid operation. Please try again.")

    try:
        num2 = float(input("Enter the second number: "))
    except ValueError:
        print("Invalid input. Please enter a numeric value.")
        continue

    result = calculate(num1, num2, choice)

    if choice == "/" and num2 == 0:
        print("\nError: Division by zero\n")
    else:
        print(f"\nResult: {num1} {choice} {num2} = {calculate(num1, num2, choice)}\n")

    again = input("Do you want to perform another calculation? (yes/no): ").strip().lower()
    if again != "yes":
        print("Thank you for using the calculator. Goodbye!")
        break
```
