```python
tform =  transform.SimilarityTransform(matrix= np.array([[1,0,10],
                                                        [0,1,10],
                                                        [0,0,1]]))

tform
im_ = transform.warp(img_camera, tform)
imshow(im_)
plt.imshow(im_)
```

```python
tform =  transform.SimilarityTransform(rotation=np.deg2rad(10))
tform
```

```python
im_ = transform.warp(img_camera, tform.inverse)
imshow(im_)
plt.show()
```

```python
theta = np.deg2rad(10)
center = np.array(img_camera.shape) //2
A = np.array([[1, 0, -center[1]],
                [0, 1, -center[0]],
                [0, 0, 1]])

B = np.array([[np.cos(theta), -np.sin(theta), 0],
                [np.sin(theta), np.cos(theta), 0],
                [0, 0, 1]])

C = np.array([[1, 0, center[1]],
                [0, 1, center[0]],
                [0,0,1]])
tform = transform.SimilarityTransform(matrix = np.dot(np.dot(C, B), A))
im_ = transform.warp(image = img_camera,
                     inverse_map = tform.inverse)
imshow(im_)
plt.show()
print(tform)
```
