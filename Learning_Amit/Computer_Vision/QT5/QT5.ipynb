```python
img_dog = cv2.imread("C:/Users/Mgama/Ai_Amit_Diploma/Learning_Amit/Computer_Vision/QT5/IMG-20251115-WA0017.jpg")
src = cv2.cvtColor(img_dog, cv2.COLOR_BGR2GRAY)
dst = cv2.equalizeHist(src)
cv2.imshow(src)
cv2.imshow(dst)
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Original Gray")
plt.imshow(src, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Histogram Equalized")
plt.imshow(dst, cmap='gray')
plt.axis('off')

plt.show()
```

```python
img =  cv2.imread('C:/Users/Mgama/Ai_Amit_Diploma/Learning_Amit/Computer_Vision/QT5/IMG-20251115-WA0017.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(gray, (3, 3), 0)

sobelX = cv2.Sobel(img_blur, cv2.CV_64F, 1, 0, ksize=5)
sobelY = cv2.Sobel(img_blur, cv2.CV_64F, 0, 1, ksize=5)

plt.imshow(img_blur, cmap='gray')
plt.title('Blurred')
plt.show()

plt.imshow(sobelX, cmap='gray')
plt.title('Sobel X')
plt.show()

plt.imshow(sobelY, cmap='gray')
plt.title('Sobel Y')
plt.show()
```

```python
img =  cv2.imread('C:/Users/Mgama/Ai_Amit_Diploma/Learning_Amit/Computer_Vision/QT5/IMG-20251115-WA0017.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(gray, (3, 3), 0)

sobelX = cv2.Sobel(img_blur, cv2.CV_64F, 1, 0, ksize=5)
sobelY = cv2.Sobel(img_blur, cv2.CV_64F, 0, 1, ksize=5)

plt.imshow(img_blur, cmap='gray')
plt.title('Blurred')
plt.show()

plt.imshow(sobelX, cmap='gray')
plt.title('Sobel X')
plt.show()

plt.imshow(sobelY , cmap='gray')
plt.title('Sobel Y')
plt.show()

plt.imshow(sobelY + sobelX , cmap='gray')
plt.title('Sobel Y + X')
plt.show()
```
