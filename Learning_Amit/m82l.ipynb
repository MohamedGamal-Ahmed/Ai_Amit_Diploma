```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np. array ([1, 2, 3, 4])
y = np.array ([2, 2.8, 3.6, 4.5])
```

```python
w = 0
b = 0
alpha = 0.01
num_iterations = 20
```

```python
sse_values = []
# GD
for i in range(num_iterations):
    y_hat = w * x + b

    # partial derivatives

    D_w = 2/4 * np.sum((y_hat - y) * x)
    D_b = 2/4 * np.sum(y_hat - y)*1

    # update parameters

    w -= alpha * D_w
    b -= alpha * D_b

    see = (np.sum((y_hat - y)**2))
    sse_values.append(see)

    if (i+1) % 20 == 0:
        print(f"Iteration {i+1}, SSE = {see}")
```

```python
print (f"optimized param : bais = {b}, weight = {w}")
```

```python
plt . figure (figsize = (12, 5))
plt.plot(range(num_iterations), sse_values, label='SSE')
```

```python
plt . figure (figsize = (12, 5))
plt.scatter(x,y , color='red', label='Data point')

plt.plot(x , w * x + b, color='blue', label='Regression Line')
```
