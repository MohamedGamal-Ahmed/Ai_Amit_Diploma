```python
import csv
def calculate_averages(input_file, output_file):
    students = []

    # step 1: Read the CSV file and  calculate averages
    with open (input_file , newline='') as csvfile:
        reader = csv.DictReader(csvfile) #dictionary convertor
        for row in reader:
            name = row['name']
            scores = [int(row['math']), int(row['english']), int(row['science'])]
            average = round(sum(scores) / len(scores) ,2) # round the average to 2 decimal places
            students.append({'name': name, 'average': average})

    # step 2 : sort by average score
    top_students = sorted(students, key=lambda x: x['average'], reverse=True)[:5]  # get top 5 students

    # step 3: Write the results to a new CSV file
    with open(output_file, 'w', newline='') as csvfile:
        
        writer = csv.DictWriter(csvfile, fieldnames=['name' , 'average'])
        writer.writeheader()
        writer.writerows(top_students)
        
# Example usage
calculate_averages('students_new .csv', 'averages.csv')


```

```python
import pandas as pd   # for dataframe
import string  # for string manipulation
import nltk
from nltk.corpus import stopwords  # for stop words

#download the stopwords if not already downloaded
nltk.download('stopwords')

#load the dataset
df = pd.read_csv("C:/Users/Mgama/Ai_Amit_Diploma/email.csv")

df.head()  # display the first few rows of the dataset


```

```python
def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    cleaned_words = [word for word in text.split() if word not in stop_words]
    return ' '.join(cleaned_words)

#apply preprocessing to the 'text' column
df['cleaned_text'] = df['text'].apply(clean_text)

#show the cleaned text
print (df[['text', 'cleaned_text']])
```
